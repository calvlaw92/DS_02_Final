```python
def apply_scale(dataframe, scaling_function):
    num_df = dataframe.select_dtypes(exclude=['category'])
    numerical_log_df = np.log(num_df + 1)
    print(numerical_log_df.columns)
    numerical_log_df = scaling_function(numerical_log_df)
    tmp_df = dataframe.copy()
    tmp_df[numerical_log_df.columns] = numerical_log_df
    return tmp_df

def gelman_scale(dataframe):
    return (dataframe - dataframe.mean())/(2*dataframe.std())

def standard_scale(dataframe):
    return (dataframe - dataframe.mean())/(dataframe.std())

df_gelman = apply_scale(df_housing_original, gelman_scale)
df_standard = apply_scale(df_housing_original, standard_scale)
```

```python
features = ['SalePrice','MasVnrArea','LotArea','FirstFlrSF']
```

```python
fig = plt.figure(figsize=(20,9))
for i, col in enumerate(features):
    fig.add_subplot(3,4,1+i)
    sns.distplot(df_housing_original[col])
    plt.axvline(df_housing_original[col].mean(), c='red')
    plt.axvline(df_housing_original[col].median(), c='black')
    plt.legend()
    fig.add_subplot(3,4,5+i)
    sns.distplot(df_gelman[col])
    plt.xlim(-3,3)
    plt.axvline(df_gelman[col].mean(), c='red')
    plt.axvline(df_gelman[col].median(), c='black')
    plt.legend()
    fig.add_subplot(3,4,9+i)
    sns.distplot(df_standard[col])
    plt.xlim(-3,3)
    plt.axvline(df_standard[col].mean(), c='red')
    plt.axvline(df_standard[col].median(), c='black')
    plt.legend()
```
